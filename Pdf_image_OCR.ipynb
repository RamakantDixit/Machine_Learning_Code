```python
from pdf2image import convert_from_path
import cv2 as cv
import numpy as np 
import pytesseract
```

```python
#read pdf file and convert it to images
images = convert_from_path(r"C:\Users\Kanan\Desktop\Wipro Python Assement\ocr\pdf\1_100A_39015.pdf")
for i,image in enumerate(images):
    filename=r"C:\Users\Kanan\Desktop\Wipro Python Assement\ocr\pdf\image_"+str(i)+".jpg"
    image.save(filename,"JPEG")                         
```

```python
# image thresholding
filename=r"C:\Users\Kanan\Desktop\Wipro Python Assement\ocr\pdf"
for i,image in enumerate(images):
    img = cv.imread(filename+"\image_"+str(i)+".jpg")
    ret,thresh1 = cv.threshold(img,127,255,cv.THRESH_BINARY_INV)
    #thresh1 = cv.adaptiveThreshold(img,125,cv.ADAPTIVE_THRESH_GAUSSIAN_C,
    #cv.THRESH_BINARY_INV,75,10) 
    cv.imwrite(filename+"\image_THRESH_BINARY_INV_"+str(i)+".jpg",thresh1)
```

```python
# remove the noice and blur an image
#this is not req always can be skiped
filename=r"C:\Users\Kanan\Desktop\Wipro Python Assement\ocr\pdf"
for i,image in enumerate(images):
    img = cv.imread(filename+"\image_THRESH_BINARY_INV_"+str(i)+".jpg")
    gausBlur = cv.GaussianBlur(img, (1,1),0)  
    cv.imwrite(filename+"\image_blur_"+str(i)+".jpg",gausBlur)
```

```python
#Morphological operations from blur
#Erosion 
kernel_e = np.ones((1,1), np.uint8) 
kernel_d = np.ones((1,1), np.uint8)
img = cv.imread(filename+"\image_blur_1.jpg")
img_erosion = cv.erode(img, kernel_e, iterations=1)
cv.imwrite(filename+"\img_erosion_1.jpg",img_erosion)

#Dilation
img = cv.imread(filename+"\img_erosion_1.jpg")
img_dilation = cv.dilate(img, kernel_d, iterations=1) 
cv.imwrite(filename+"\img_dilation_1.jpg",img_dilation)
```
